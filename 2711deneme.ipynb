```python
import json

input_file = "dataset.jsonl"
output_file = "dataset_fixed.jsonl"

with open(input_file, "r", encoding="utf-8") as f_in, open(output_file, "w", encoding="utf-8") as f_out:
    for i, line in enumerate(f_in, start=1):
        if not line.strip():
            continue
        
        obj = json.loads(line)
        
        # Eğer input alanı dict ise → string'e çevir
        if isinstance(obj.get("input"), dict):
            obj["input"] = json.dumps(obj["input"], ensure_ascii=False)
        
        # Eğer instruction veya output dict ise aynı şekilde düzelt
        if isinstance(obj.get("instruction"), dict):
            obj["instruction"] = json.dumps(obj["instruction"], ensure_ascii=False)
        if isinstance(obj.get("output"), dict):
            obj["output"] = json.dumps(obj["output"], ensure_ascii=False)
        
        f_out.write(json.dumps(obj, ensure_ascii=False) + "\n")

print(" Düzenleme tamamlandı → dataset_fixed.jsonl oluşturuldu.")
```

```python
#tekrar çalıştırmaya gerek yok fine tuning edilmiş model artık elimizde
import os
import json
import torch
from datasets import load_dataset
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    TrainingArguments,
    Trainer,
)
from peft import LoraConfig, get_peft_model

# =========================================================
# 1) CUDA KONTROL
# =========================================================
print("CUDA:", torch.cuda.is_available(), "-", torch.version.cuda)
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))

# =========================================================
# 2) MODEL + TOKENIZER (4-bit QLoRA)
# =========================================================
model_id = "Qwen/Qwen2.5-7B-Instruct"

bnb = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
)

tokenizer = AutoTokenizer.from_pretrained(model_id, use_fast=True)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# === EN KRİTİK DÜZELTME ===
# - torch_dtype KALKTI
# - device_map="auto" eklendi
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb,
    device_map="auto",
    low_cpu_mem_usage=True,
)

# =========================================================
# 3) LoRA
# =========================================================
lora_cfg = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],
    task_type="CAUSAL_LM",
)
model = get_peft_model(model, lora_cfg)
print("4-bit QLoRA hazır.")

# =========================================================
# 4) DOSYALAR
# =========================================================
DATASET_MAIN = "dataset_fixed.jsonl"
DATASET_PROMPT = "promt_base..txt"
DATASET_TABLES = "tables.json"

assert os.path.exists(DATASET_MAIN)
assert os.path.exists(DATASET_PROMPT)
assert os.path.exists(DATASET_TABLES)

with open(DATASET_PROMPT, "r", encoding="utf-8") as f:
    SYSTEM_PROMPT = f.read().strip()

raw = load_dataset("json", data_files={"data": DATASET_MAIN})["data"]
ds = raw.train_test_split(test_size=0.05)

# =========================================================
# 5) TOKENIZE
# =========================================================
MAX_LEN = 1024

def tokenize(ex):
    text = (
        f"<|im_start|>system\n{SYSTEM_PROMPT}<|im_end|>\n"
        f"<|im_start|>user\n"
        f"Soru: {ex['instruction']}\n"
        f"Şema: {ex['input']}\n"
        f"Sadece PostgreSQL uyumlu tam SQL üret.\n"
        f"<|im_end|>\n"
        f"<|im_start|>assistant\n{ex['output']}<|im_end|>"
    )

    enc = tokenizer(
        text,
        truncation=True,
        max_length=MAX_LEN,
        padding=False,
        add_special_tokens=False,
    )

    ids = enc["input_ids"]
    return {
        "input_ids": ids,
        "attention_mask": [1] * len(ids),
        "labels": ids.copy(),
    }

train = ds["train"].map(tokenize, remove_columns=ds["train"].column_names)
valid = ds["test"].map(tokenize, remove_columns=ds["test"].column_names)

print("Train:", len(train), "Valid:", len(valid))

# =========================================================
# 6) MEMORY OPTIMIZATION
# =========================================================
model.gradient_checkpointing_enable()
model.enable_input_require_grads()
model.config.use_cache = False

# =========================================================
# 7) COLLATE
# =========================================================
def collate(batch):
    pad = tokenizer.pad_token_id
    maxlen = max(len(x["input_ids"]) for x in batch)

    def pad_to(seq, fill):
        return seq + [fill] * (maxlen - len(seq))

    return {
        "input_ids": torch.tensor([pad_to(x["input_ids"], pad) for x in batch]),
        "attention_mask": torch.tensor([pad_to(x["attention_mask"], 0) for x in batch]),
        "labels": torch.tensor([pad_to(x["labels"], -100) for x in batch]),
    }

# =========================================================
# 8) TRAINING — Evaluation kapalı
# =========================================================
args = TrainingArguments(
    output_dir="sqlcoder-lora",
    num_train_epochs=2,
    per_device_train_batch_size=1,
    gradient_accumulation_steps=16,
    learning_rate=1e-4,
    warmup_ratio=0.03,
    logging_steps=20,

    evaluation_strategy="no",
    save_strategy="epoch",

    bf16=True,
    optim="adamw_torch",
    report_to="none",
    gradient_checkpointing=True,
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train,
    eval_dataset=None,
    data_collator=collate,
    tokenizer=tokenizer,
)

# =========================================================
# 9) TRAIN 
# =========================================================
trainer.train()

# =========================================================
# 9) SAVE
# =========================================================
model.save_pretrained("sqlcoder-lora")
tokenizer.save_pretrained("sqlcoder-lora")

print("Eğitim tamamlandı → sqlcoder-lora") # klasörünün içine kaydeder fine tuning edilmiş model artık buradadır dosyaları birleştirip veya ayrı ayrı da kullanılabilir rest api şeklinde ya da ollamaya entegre edilmiş halde ya da hugging face yükleyip oradan modeli token ile çekip kullanılabilir.

# Bu model fine tuning edilmiştir yani her hangi bir güncellemede tekrardan bu modeli veri setine veya promta göreı ekleme çıkarma yaparak eğitmek gerekir

# rag ile veriyi cuhunklara ayırarak modeli eğitirsek sıfırdan bir eğitime gerek kalmaz ve güncellenen veri setinde ya da herhangi bir gelişmede kolayca güncelleyebiiliriz ama rag sistemi mantıklı değildir maliyet ve çıktı olarak fine tuning kadar ucuz ve doğru sonuç vermez
```

```python
# =========================================================
# 0) IMPORTS
# =========================================================
import os
import json
import re
import torch
from datasets import load_dataset
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig,
    TrainingArguments,
    Trainer,
    EarlyStoppingCallback,
)
from peft import LoraConfig, get_peft_model


# =========================================================
# 0.1) SQL TIRNAKLAMA FONKSİYONLARI
# =========================================================

def quote_if_needed(name: str) -> str:
    """
    PostgreSQL kuralı:
    - Sadece küçük harf, rakam, alt çizgi → tırnaksız
    - Diğer tüm durumlar → çift tırnak
    """
    if re.fullmatch(r"[a-z0-9_]+", name):
        return name
    return f"\"{name}\""


def preprocess_sql(sql: str) -> str:
    """
    SQL içindeki tablo/sütun isimlerini bulup tırnak gereksinimini uygular.
    Basit ama etkili bir normalizasyon: \bidentifier\b yakalar.
    """
    tokens = re.findall(r"[A-Za-z0-9_şŞğĞüÜİıÖöÇç]+", sql)

    for t in tokens:
        fixed = quote_if_needed(t)
        sql = re.sub(rf"\b{t}\b", fixed, sql)

    return sql


# =========================================================
# 1) CUDA KONTROLÜ
# =========================================================
print("CUDA:", torch.cuda.is_available(), "-", torch.version.cuda)
if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))


# =========================================================
# 2) MODEL + TOKENIZER (4-bit QLoRA yükleme)
# =========================================================
model_id = "Qwen/Qwen2.5-7B-Instruct"

bnb = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
)

tokenizer = AutoTokenizer.from_pretrained(model_id, use_fast=True)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb,
    device_map="auto",
    low_cpu_mem_usage=True,
)


# =========================================================
# 3) LoRA KONFİGÜRASYONU
# =========================================================
lora_cfg = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],
    task_type="CAUSAL_LM",
)

model = get_peft_model(model, lora_cfg)
print("4-bit QLoRA hazır.")


# =========================================================
# 4) DOSYALAR
# =========================================================
DATASET_MAIN = "dataset_fixed.jsonl"
DATASET_PROMPT = "promt_base..txt"
DATASET_TABLES = "tables.json"

assert os.path.exists(DATASET_MAIN)
assert os.path.exists(DATASET_PROMPT)
assert os.path.exists(DATASET_TABLES)

with open(DATASET_PROMPT, "r", encoding="utf-8") as f:
    SYSTEM_PROMPT = f.read().strip()

raw = load_dataset("json", data_files={"data": DATASET_MAIN})["data"]
ds = raw.train_test_split(test_size=0.05)


# =========================================================
# 5) TOKENIZATION PIPELINE
# =========================================================
MAX_LEN = 1024

def tokenize(ex):
    # === SQL çıktısını normalize et (tırnak kuralı) ===
    sql = preprocess_sql(ex["output"])

    # Template
    text = (
        f"<|im_start|>system\n{SYSTEM_PROMPT}<|im_end|>\n"
        f"<|im_start|>user\n"
        f"Soru: {ex['instruction']}\n"
        f"Şema: {ex['input']}\n"
        f"Sadece PostgreSQL uyumlu tam SQL üret.\n"
        f"<|im_end|>\n"
        f"<|im_start|>assistant\n{sql}<|im_end|>"
    )

    enc = tokenizer(
        text,
        truncation=True,
        max_length=MAX_LEN,
        padding=False,
        add_special_tokens=False,
    )

    ids = enc["input_ids"]

    return {
        "input_ids": ids,
        "attention_mask": [1] * len(ids),
        "labels": ids.copy(),
    }


train = ds["train"].map(tokenize, remove_columns=ds["train"].column_names)
valid = ds["test"].map(tokenize, remove_columns=ds["test"].column_names)

print("Train:", len(train), "Valid:", len(valid))


# =========================================================
# 6) BELLEK OPTİMİZASYONU
# =========================================================
model.gradient_checkpointing_enable()
model.enable_input_require_grads()
model.config.use_cache = False


# =========================================================
# 7) COLLATE (Dinamik padding)
# =========================================================
def collate(batch):
    pad = tokenizer.pad_token_id
    maxlen = max(len(x["input_ids"]) for x in batch)

    def pad_to(seq, fill):
        return seq + [fill] * (maxlen - len(seq))

    return {
        "input_ids": torch.tensor([pad_to(x["input_ids"], pad) for x in batch]),
        "attention_mask": torch.tensor([pad_to(x["attention_mask"], 0) for x in batch]),
        "labels": torch.tensor([pad_to(x["labels"], -100) for x in batch]),
    }


# =========================================================
# 8) TRAINING ARGUMENTS
# =========================================================
args = TrainingArguments(
    output_dir="sqlcoder-lora",
    num_train_epochs=2,
    per_device_train_batch_size=1,
    gradient_accumulation_steps=8,
    learning_rate=5e-5,
    warmup_ratio=0.08,
    logging_steps=20,
    evaluation_strategy="steps",
    eval_steps=40,
    save_strategy="steps",
    save_steps=40,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    bf16=True,
    optim="adamw_torch",
    report_to="none",
    gradient_checkpointing=True,
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train,
    eval_dataset=valid,
    data_collator=collate,
    tokenizer=tokenizer,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=3)],
)


# =========================================================
# 9) EĞİTİM + KAYDETME
# =========================================================
SAVE_DIR = "checkpoints/sqlcoder-lora-v2"

model.save_pretrained(SAVE_DIR)
tokenizer.save_pretrained(SAVE_DIR)

print("Eğitim tamamlandı →", SAVE_DIR)
```

```python
#kontrol aşaması
import json
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel

# ===============================
# 1) Model Yükleme
# ===============================
BASE_MODEL = "Qwen/Qwen2.5-7B-Instruct"
LORA_DIR = "sqlcoder-lora"
TEST_PATH = "test_500.jsonl"

bnb = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
)

tokenizer = AutoTokenizer.from_pretrained(BASE_MODEL, use_fast=True)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

base = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL,
    quantization_config=bnb,
    device_map="auto",
    low_cpu_mem_usage=True,
)

model = PeftModel.from_pretrained(base, LORA_DIR, device_map="auto")
model.eval()

print("Model yüklendi")

# ===============================
# 2) Test seti yükleme
# ===============================
tests = []
with open(TEST_PATH, "r", encoding="utf8") as f:
    for line in f:
        if line.strip():
            tests.append(json.loads(line))

print("Test örneği:", len(tests))

# ===============================
# 3) Minimal Prompt (Sadece SQL Üret)
# ===============================
def build_prompt(ex):
    return (
        "<|im_start|>system\n"
        "Sadece PostgreSQL uyumlu tam SQL üret. Açıklama verme. Yorum yok. Sadece SQL.\n"
        "<|im_end|>\n"
        f"<|im_start|>user\n{ex['instruction']}\n<|im_end|>\n"
        "<|im_start|>assistant\n"
    )

# ===============================
# 4) 500 TEST + TXT Kaydetme
# ===============================
OUT_FILE = "test_output3.txt"
with open(OUT_FILE, "w", encoding="utf-8") as f:
    f.write("==== SQLCODER SADE ÇIKTI (Soru + SQL) ====\n\n")

for i, ex in enumerate(tests):
    prompt = build_prompt(ex)
    enc = tokenizer(prompt, return_tensors="pt").to(model.device)

    out = model.generate(
        **enc,
        max_new_tokens=200,
        do_sample=False,
        temperature=0.0
    )

    sql_text = tokenizer.decode(out[0], skip_special_tokens=True).strip()

    print(f"[{i+1}/500] Tamamlandı")

    with open(OUT_FILE, "a", encoding="utf-8") as f:
        f.write(f"Soru: {ex['instruction']}\n")
        f.write(f"SQL: {sql_text}\n\n")

print("\n 500 test işlendi ve 'test_output3.txt' dosyasına kaydedildi.")
```
